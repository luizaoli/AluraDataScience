1. Calcular a média de quartos por apartamentos.

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Para exibição mais bonita no Jupyter
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

# Carregando o dataset
arquivo = "dados_aluguel-2.csv"
dados = pd.read_csv(arquivo, sep=";")
dados.head()

# Estrutura básica do dataset
print("Dimensão do dataset (linhas, colunas):", dados.shape)
print("\nNome das colunas:")
print(dados.columns)

# Informações gerais sobre tipos de dados
print("\nInformações do DataFrame:")
dados.info()

# Calculando a média de quartos considerando apenas imóveis do tipo "Apartamento"

# Filtrando apenas os imóveis do tipo "Apartamento"
apartamentos = dados[dados["Tipo"] == "Apartamento"]

media_quartos_apartamento = apartamentos["Quartos"].mean()

print(f"Média de quartos por apartamentos: {media_quartos_apartamento:.2f}")
```

3. Analisar quais bairros possuem a media de valor de aluguel mais elevado.

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Para exibição mais bonita no Jupyter
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

# Carregando o dataset
arquivo = "dados_aluguel-2.csv"
dados = pd.read_csv(arquivo, sep=";")
dados.head()

# Estrutura básica do dataset
print("Dimensão do dataset (linhas, colunas):", dados.shape)
print("\nNome das colunas:")
print(dados.columns)

# Informações gerais sobre tipos de dados
print("\nInformações do DataFrame:")
dados.info()

# Bairros com as maiores médias de valor de aluguel

# Agrupando por bairro e calculando a média do valor de aluguel
media_valor_por_bairro = (
    dados.groupby("Bairro")["Valor"]
         .mean()
         .sort_values(ascending=False)
)

print("Bairros com as maiores médias de valor de aluguel:")
media_valor_por_bairro.head(10)
```

4. Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevados.

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Para exibição mais bonita no Jupyter
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

# Carregando o dataset
arquivo = "dados_aluguel-2.csv"
dados = pd.read_csv(arquivo, sep=";")
dados.head()

# Estrutura básica do dataset
print("Dimensão do dataset (linhas, colunas):", dados.shape)
print("\nNome das colunas:")
print(dados.columns)

# Informações gerais sobre tipos de dados
print("\nInformações do DataFrame:")
dados.info()

# Criando o gráfico de barras horizontais - Top 5 bairros por média de aluguel

top5_bairros = media_valor_por_bairro.head(5)

plt.figure(figsize=(10, 6))

# barh = gráfico de barras horizontais
plt.barh(top5_bairros.index, top5_bairros.values)

plt.xlabel("Média do valor de aluguel (R$)")
plt.ylabel("Bairro")
plt.title("Top 5 bairros com maior média de valor de aluguel")

# Inverter a ordem no eixo Y para o maior ficar em cima
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

5. Identificar quais apartamentos possuem área maior que 80 m2 ou aluguel menor que R$ 4.000,00.

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Para exibição mais bonita no Jupyter
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

# Carregando o dataset
arquivo = "dados_aluguel-2.csv"
dados = pd.read_csv(arquivo, sep=";")
dados.head()

# Estrutura básica do dataset
print("Dimensão do dataset (linhas, colunas):", dados.shape)
print("\nNome das colunas:")
print(dados.columns)

# Informações gerais sobre tipos de dados
print("\nInformações do DataFrame:")
dados.info()

# Identificando os apartamentos com área maior que 80 m² OU aluguel menor que R$ 4.000,00
filtro = (
    (dados["Tipo"] == "Apartamento") &
    ((dados["Area"] > 80) | (dados["Valor"] < 4000))
)

apartamentos_filtrados = dados[filtro]

print("Quantidade de apartamentos que atendem ao critério:", len(apartamentos_filtrados))
apartamentos_filtrados.head(20)
```
